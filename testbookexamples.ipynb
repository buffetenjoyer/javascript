```typescript
let countTheAs = (...grades) => {
    let count = 0;
    for (let grade of grades) {
        if (grade == "A") {
            count++;
        }
    }
    return count;
}

console.log(countTheAs("A", "B+", "A-", "A", "D", "C", "C", "C-"));

let gradeBook = ["A", "B+", "A-", "A", "D", "C", "C", "C-"];

console.log(countTheAs(gradeBook));              // doesn't work properly, returns 0, when we want to it to return 2

console.log(countTheAs(...gradeBook));           // have to use ... to spread the array into seperate entries
```

```typescript
function repeat(times, action) {
    for (let i = 0; i < times; i++) {
        action(i);
    }
}

repeat(5,console.log);    // works, as a function value itself is being input to the repeat function
repeat(5,console.log());  // doesn't work, as you are inputting the return value of console.log (not its function value itself) by adding those parenthesis after console.log
```

```typescript
function experiment(inputSet, func, trials = 1) {
    return () => {
        console.log("Experiment");
        for (let i = 1; i <  trials+1; i++) {
            let str = `trial ${i}:-------------`;           // have to use backtick for the ${} embed mechanic to work, backtick is ` , not ' . The latter is apostrophe/single quote.
            console.log(str);
            for (let input of inputSet) {
                console.log(`input: ${input},  output: ${func(input)}` );
            }
        }

    }   
}

let printExperiment = experiment([1,2,3], (x) => {return x+1}, 3);       // remember you need statement inside braces for short function notation, in other words you need to write "return x+1", not just "x+1" if you use braces

printExperiment();

let randomnessExperiment = experiment([1,2,3], (x) => {return x+Math.random()}, 3)

randomnessExperiment();

let guessRandomNumberExperiment = experiment([1,2,3], (x) => {
    let guess = -1;
    let guesses = [];
    while (guess != x) {
        guess = Math.round(Math.random()*10);
        guesses.push(guess)
    }
    return guesses;    
}, 3);

guessRandomNumberExperiment();

let guessRandomNumberExperiment1Trial = experiment([1,2,3], (x) => {
    let guess = -1;
    let guesses = [];
    while (guess != x) {
        guess = Math.round(Math.random()*10);
        guesses.push(guess)
    }
    return guesses;    
});

guessRandomNumberExperiment1Trial();
```
