```typescript
//     Chapter 6 Excercise 1


class Vec {
    constructor(x, y) {
        this.x = x;
        this.y = y;
    }

    plus(Vec2) {
        return new Vec(this.x + Vec2.x, this.y + Vec2.y);
    }

    minus(Vec2) {
        return new Vec(this.x - Vec2.x, this.y - Vec2.y);
    }

    get length() {
        return Math.sqrt(this.x*this.x + this.y*this.y);
    }
}

let vector1 = new Vec(1, 4);
let vector2 = new Vec(2,-3);

console.log(vector1.plus(vector2));
console.log(vector1.minus(vector2));

console.log(vector1.minus(vector2).length);
console.log(vector2.length);
```

```typescript
function createSetOfSimulations(graph, numOfParcels, numOfSimulations) {
    let Simulations = []
    for (let i = 0; i < numOfSimulations; i++) {
        //console.log("hu");
        let parcels = [];
        for (let i = 0; i < numOfParcels; i++) {
            let randomVertice = randomPick(graph);
            let randomPlace = randomVertice.place;
            let randomDeliveryAddress = randomPlace;
            while (randomDeliveryAddress == randomPlace) {
                randomDeliveryAddress = randomPick(graph).place;
            }
            let parcel = {location: randomPlace, deliveryAddress: randomDeliveryAddress};
            parcels.push(parcel);
        }
        let startingLocation = randomPick(graph).place; 
        //console.log(startingLocation);
        let state = new villageState(startingLocation, parcels); 
        //console.log(state);
        let simulation = new Simulation(state, graph);
        Simulations.push(simulation);
    }
    return Simulations;
   
}

function runTheSimulations(Simulations, algorithm) {
    let copySimulations = Simulations.map(simulation => {
        let copyParcels = [];
        for (let parcel of simulation.villageState.parcels) {
            let copyParcel = {location: parcel.location, deliveryAddress: parcel.deliveryAddress}
            copyParcels.push(copyParcel);
        }
        let copy = new Simulation(new villageState(simulation.villageState.robotLocation, 
        copyParcels), simulation.graph);
        return copy;
    }
    );
    
    let trackingAverage = 0;
    let totalSteps = 0;
    for (let i = 0; i < copySimulations.length; i++) {
        let steps = playTheSimulation(copySimulations[i], algorithm);
        totalSteps+= steps;
        trackingAverage = totalSteps/(i+1); 
        console.log("current Average: " + trackingAverage);
    }
    
    return trackingAverage;
};
let inputSimulations = createSetOfSimulations(villageGraph, 5, 5);
let finalAverage = runTheSimulations(inputSimulations, randomAlgorithm);
console.log(finalAverage);
console.log("------------------------------------------------------------------");
let finalAverage2 = runTheSimulations(inputSimulations, breadthFirstSearchAlgorithm);
console.log(finalAverage2);


function breadthFirstSearchAlgorithm(state, graph, memory) {
    if (memory.places == undefined) {
        memory.places = [];
        memory.currentRoute = [];
        memory.places.push({location: state.robotLocation, route: [state.robotLocation], checked: false});
    }
    if (memory.currentRoute.length != 0) {
        let next = memory.currentRoute[0];
        memory.currentRoute = memory.currentRoute.slice(1);
        return (next);
    }
    let lookingFor;
    if (state.parcels[0].location == state.robotLocation) {
        lookingFor = state.parcels[0].deliveryAddress;
    }
    else {
        lookingFor = state.parcels[0].location;
    }
    if (memory.places.some(memoryPiece => memoryPiece.location == lookingFor)) {
        let foundMemory = memory.places.find(memoryPiece => memoryPiece.location == lookingFor);
        console.log("made it here not");
        memory.currentRoute = foundMemory.route.slice(1).concat(foundMemory.route.slice(0, foundMemory.route.length - 1).reverse());
        console.log("made it here");
    }
    else {
        for (let i = 0; state.parcels.length != 0;) {
            console.log(memory.places);
            let dequeued = memory.places.find(memoryPiece => !memoryPiece.checked);
            if (dequeued.location == lookingFor) {
                console.log("made it not");
                console.log(dequeued.route);
                console.log(dequeued.route.slice(1));
                memory.currentRoute = dequeued.route.slice(1).concat(dequeued.route.slice(0, dequeued.route.length - 1).reverse());
                console.log(memory.currentRoute);
                console.log("made it");
                break;
            }
            let dequeuedVertice = graph.find(vertice => vertice.place == dequeued.location);
            let edges = dequeuedVertice.edges;
            for (let connectedLocation of edges) {
                if (!memory.places.some(memoryPeice => memoryPeice.location == connectedLocation)) {
                    console.log(connectedLocation);
                    memory.places.push({location: connectedLocation, route: dequeued.route.concat(connectedLocation), checked: false})
                    console.log("here2");
                }    
            }
            dequeued.checked = true;
        }
    }
    if (memory.currentRoute.length != 0) {
        console.log("dede");
        let next = memory.currentRoute[0];
        console.log("dada");
        memory.currentRoute = memory.currentRoute.slice(1);
        return next;
    }
}
```
