```typescript
// Chapter 2 Excercise 3 part 1

let checkerboard = "";

for (let rowPair = 1; rowPair <= 4; rowPair++) {                     // add pair of rows (two rows) per loop
    for (let tilePair = 1; tilePair <= 4; tilePair++) {              // first row of pair
        checkerboard += " #";
    }
    checkerboard += "\n";
    for (let tilePair = 1; tilePair <= 4; tilePair++) {             // second row of pair
        checkerboard += "# ";
    }
    if (rowPair == 4) {
        break;
    }
    checkerboard += "\n";
}

console.log(checkerboard);
```

```typescript
// Chapter 2 Excercise 3 part 2

let size = 10;
let checkerboard2 = "";
if (size != 0) {
    for (let row = 1; row <= size; row++) {
        if (row %2 == 1) {
            for (let tile = 1; tile <= size; tile++) {
                if (tile%2 == 1) {
                    checkerboard2 += " ";
                }
                else {
                    checkerboard2 += "#";
                }
            }
        }
        else {
            for (let tile = 1; tile <= size; tile++) {
                if (tile%2 == 1) {
                    checkerboard2 += "#";
                }
                else {
                    checkerboard2 += " ";
                }
                
            }
        }
        if (row != size) {
            checkerboard2 += "\n";
        }
    }
}


console.log(checkerboard2);
```

```typescript
// Chapter 3 Excercise 2

let isEven = function(n) {
    if (n == 0) {
        return true;
    }
    if (n == 1) {
        return false;
    }
    if (n < 0) {
        return isEven(n + 2);
    }
    else {
        return isEven(n - 2);
    }
}
console.log(isEven(0));
console.log(isEven(1));
console.log(isEven(2));
console.log(isEven(13));
console.log(isEven(50));
console.log(isEven(75));
console.log(isEven(-1));
console.log(isEven(-110));
```

```typescript
// Chapter 3 Excercise 3

let countBs = function(string1) {
    let count = 0;
    for (let i = 0; i < string1.length; i++) {
        if (string1[i] == "B") {
            count++;
        }
    }
    return count;
}

let countChar = function(string1, char1) {
    let count = 0;
    for (let i = 0; i < string1.length; i++) {
        if (string1[i] == char1) {
            count++;
        }
    }
    return count;
}

let countBs2 = function(string1) {
    return countChar(string1, "B");
}

console.log(countBs("Binary"));
console.log(countBs(""));
console.log(countBs("Happy"));
console.log(countBs("BoBBy"));
console.log(countBs("bigBen"));

console.log(countBs2("Binary"));
console.log(countBs2(""));
console.log(countBs2("Happy"));
console.log(countBs2("BoBBy"));
console.log(countBs2("bigBen"));
```

```typescript
//  Chapter 4 Excercise 1

// part 1

function range(start, end) {
    let result = [];
    for (let i = start; i <= end; i++) {
        result.push(i);
    }
    return result;
}

console.log(range(17,30));


// part 2

function sum(arrayOfNum) {
    let summation = 0;
    for (let input of arrayOfNum) {
        summation += input;
    }
    return summation;
}

console.log(sum(range(1,10)));


// part 3

function range2(start, end, step = 1) {
    let result = [];
    if (step > 0) {
        for (let i = start; i <= end; i+= step) {
            result.push(i);
        }
    }
    else if (step < 0) {
        for (let i = start; i >= end; i+= step) {
            result.push(i);
        }
    }
    return result;
}


console.log(range2(10, 20, 2));

console.log(range2(33, 41));

console.log(range2(5, -15, -2));

console.log(range2(5, 10, 0));

console.log(range2(5, 2, -1));
```
