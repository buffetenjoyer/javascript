```typescript
const Person1 = {firstName: "Ali", lastName: "Baba"}
function sayHello(randomPerson) {
    console.log("Hi, " + randomPerson.firstName + " " + randomPerson.lastName + ".");
}
sayHello(Person1);
```

```typescript
interface Person {
    firstName: string;
    lastName: string;         // word before colon is variable name; word after the colon is the data type;
}
const Person2 = {firstName: "Alin", lastName: "Baban"}
function sayHelloBetter(thing: Person) {              // word before colon is variable name; word after the colon is the data type, which is Person Object;
    console.log("Hi, " + thing.firstName + " " + thing.lastName + ".");
}
sayHelloBetter(Person2);
```
